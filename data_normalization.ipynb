```python
# Dependencies
import pandas as pd
import numpy as np
```

#### Genre

```python
# Read superhero_cinema.csv as Pandas DataFrame
movie_df = pd.read_csv('Output/superhero_cinema.csv')
```

```python
genre_split_df = movie_df.assign(Genre=movie_df['Genre'].str.split(','))
genre_split_df.head()
```

```python
genre_explode_df = genre_split_df.explode('Genre').reset_index(drop=True)
genre_explode_df.head()
```

```python
# Create list of genres
genres = genre_explode_df['Genre'].unique()
len_genres = len(genres)
genre_ids = np.arange(1, len_genres + 1)
print(genres)
print(genre_ids)
```

```python
genre_id_df = pd.DataFrame({'Genre ID': genre_ids, 'Genre': genres})
genre_id_df
```

```python
genre_mapping = genre_id_df.set_index('Genre')['Genre ID'].to_dict()
genre_mapping
```

```python
genre_explode_df['Genre ID'] = genre_explode_df['Genre'].map(genre_mapping)
genre_explode_df.head(10)
```

```python
genre_relation_df = genre_explode_df[['Film ID', 'Genre ID']]
genre_relation_df.head()
```

```python
genre_explode_df.columns
```

```python
updated_movie_df = genre_explode_df[['Film ID', 'Title', 'Year', 'Rated', 'Released', 'Runtime (min)',
                                     'Director', 'Writer', 'Actors', 'Rotten Tomatoes Rating (%)',
                                     'Metascore', 'IMDB Rating', 'Box Office ($)', 'Franchise']]
updated_movie_df.drop_duplicates(inplace=True)
```

```python
updated_movie_df.to_csv('Output/superhero_no_genre.csv', index=False)
genre_id_df.to_csv('Output/genre.csv', index=False)
genre_relation_df.to_csv('Output/genre_relation.csv', index=False)
```

#### Director

```python
# Read superhero_no_genre.csv
movie_df = pd.read_csv('Output/superhero_no_genre.csv')
```

```python
# Split directors by comma
director_split_df = movie_df.assign(Director=movie_df['Director'].str.split(','))
```

```python
director_explode_df = director_split_df.explode('Director').reset_index(drop=True)
```

```python
# Create list of directors
directors = director_explode_df['Director'].unique()
len_directors = len(directors)
director_ids = np.arange(1, len_directors + 1)
print(directors)
print(director_ids)
```

```python
# Director Table
director_df = pd.DataFrame({'Director ID': director_ids, 'Director': directors})
director_df
```

```python
director_mapping = director_df.set_index('Director')['Director ID'].to_dict()
director_mapping
```

```python
director_explode_df['Director ID'] = director_explode_df['Director'].map(director_mapping)
```

```python
director_relation_df = director_explode_df[['Film ID', 'Director ID']]
director_relation_df
```

```python
director_explode_df.columns
```

```python
updated_movie_df = director_explode_df[['Film ID', 'Title', 'Year', 'Rated', 'Released', 'Runtime (min)',
                                        'Writer', 'Actors', 'Rotten Tomatoes Rating (%)',
                                        'Metascore', 'IMDB Rating', 'Box Office ($)', 'Franchise']]
updated_movie_df.drop_duplicates(inplace=True)
```

```python
updated_movie_df.to_csv('Output/superhero_no_genre_director.csv', index=False)
director_df.to_csv('Output/director.csv', index=False)
director_relation_df.to_csv('Output/director_relation.csv', index=False)
```
